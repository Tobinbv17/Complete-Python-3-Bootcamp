```python
pos = [' ']*9
def game_board(pos):
  print(pos[0]+'|'+pos[1]+'|'+pos[2])
  print(pos[3]+'|'+pos[4]+'|'+pos[5])
  print(pos[6]+'|'+pos[7]+'|'+pos[8])
```

```python
game_board('p'*10)
```

```python
def player_input():

  marker = ''
  while marker not in ['X','0'] :

    marker = input('Player1 choose marker X or O : ').upper()

    if marker == 'X':
      return ('X','O')

    elif marker == 'O':
      return ('O','X')

    else:
      print('Invalid input, choose X or O')
```

```python
#Check if the game is won,tied, lost, or ongoing.

def won_check(input_list):
  won = False
  tie = False
  ongoing = False

  won_combo = [(0, 1, 2), (3, 4, 5), (6, 7, 8), # Rows
               (0, 3, 6), (1, 4, 7), (2, 5, 8), # Columns
               (0, 4, 8), (2, 4, 6)             # Diagonals
              ]

  for combo in won_combo:

    if input_list[combo[0]] == input_list[combo[1]] == input_list[combo[2]] != ' ':
      won = True
      return (won, tie, ongoing)

  if ' ' not in input_list:
    tie = True
    return (won, tie, ongoing)

  else:
    ongoing = True
    return (won,tie,ongoing)
```

```python
#Place their input on the board.

def visualize_input() :

### player 1
  player1_marker, player2_marker = player_input()
  board = [' ']*9
  game_running = True


  while game_running == True:
    pos1 = None

    while pos1 is None:
      pos1_str = input('player1 input location 0 to 8 : ')

      if pos1_str.isdigit():
        location = int(pos1_str)

        if 0 <= location < 9:

          if board[location ] == ' ':
            pos1 = location

          else:
            print(f'Error: position taken : {board[location]}')

        else:
          print('Error: Input must be in the range 0 to 8.')

      else:
        print('Error: Invalid input. Please enter a number 0-8.')

    board[pos1] = player1_marker
    game_board(board)

    ## player1 won_check
    won,tie,game_running = won_check(board)
    if won:
      print('player1 won the game')
      break
    if tie:
      print('the game tied, nobody won!')
      break

  ### player 2
    pos2 = None
    while pos2 is None:
      pos2_str = input('player2 input location 0 to 8 : ')

      if pos2_str.isdigit():
        location = int(pos2_str)

        if 0 <= location < 9:

          if board[location ] == ' ':
            pos2 = location
          else:
            print(f'Error: position taken : {board[location]}')

        else:
          print('Error: Input must be in the range 0 to 8.')

      else:
        print('Error: Invalid input. Please enter a number 0-9.')

    board[pos2] = player2_marker
    game_board(board)

    ## player2 woncheck
    won,tie,game_running = won_check(board)
    if won:
      print('player2 won the game!')
      break
    if tie:
      print('the game tied, nobody won')
      break

  while game_running == False:
    play_again = input('Do you want to play again Y or N').upper()
    if play_again == 'Y':
      game_running = True
    else:
      print('See you next time!')
      break
```
